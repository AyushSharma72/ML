<h1>categorical data </h1>

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./Data_files/titanic.csv')
df.head()
```

<h2>1. countplot</h2> 


```python
sns.countplot(x='Survived', data=df)  # plot as a bar graph -> the frequency count of all the categories of a particular column 
df['Survived'].value_counts() # get the values
```

```python
sns.countplot(x='Pclass',data = df)
df['Pclass'].value_counts()
```

<h2>pie chart</h2>

```python
df['Survived'].value_counts().plot(kind="pie",autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind="pie",autopct="%.2f")
```

<h1>Numerical data </h1>

<h2>histogram</h2>

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=20)
```

<h2>Distplot</h2>

```python
# sns.displot(df['Age'],bins=10)
sns.displot(df['Age'], kde=True,stat='probability',color='red')# adds a smooth curve over your histogram that estimates the probability 
# density function (PDF) of your data.
```

<h2>Box Plot</h2>

```python
sns.boxplot(x='Age',data= df)  # used to identify noise and outliers in the data  it give 5 tuple summary 
```

```python
sns.boxplot(x='Fare',data=df) 
```

```python
print("Min Age:", df['Age'].min())
print("Max Age:", df['Age'].max())
print("Mean Age:", df['Age'].mean())
```

```python
print("Min Fare",df["Fare"].min())
print("Max Fare",df['Fare'].max())
print("Mean Fare:", df['Fare'].mean())
```
