```python
sns.barplot(x='Pclass', y='Fare', hue='Sex',data=titanic)
```

```python
sns.barplot(x='Sex', y='Survived',data=titanic)
```

<h2>BoxPlot (Numerical - Categorical)</h2>

```python
sns.boxplot(x='Sex',y='Age', hue='Survived', data=titanic)
```

<h2>DistPlot (Numerical - Categorical)</h2>

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label='Did not Survive')
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], label='Survived')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()

```

<h2>HeatPLot (Categorical - Categorical)</h2>

```python
titanic
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))  # to compare two categorical column
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100)
# numeric_only=True tells Pandas to ignore non-numeric columns (like names, tickets, etc.).
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind="bar")
```
