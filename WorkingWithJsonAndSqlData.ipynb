# Use SQLAlchemy with Pandas

```python
!pip install sqlalchemy pymysql
```

```python
import pandas as pd
from sqlalchemy import create_engine

username = "root"        # default XAMPP username
password = ""            # default XAMPP password is empty unless you set one
host = "localhost"
port = 3306
database = "world"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Run your query
df = pd.read_sql_query("SELECT * FROM city WHERE CountryCode ='IND'AND District='Madhya Pradesh'", engine)
print(df)
```
